```python
import re

ls = []

def word_categorize(text):
    if isinstance(text, str):
        # lowercase + remove punctuation using regex
        text = text.lower()
        text = re.sub(r'[^\w\s]', '', text)  # remove commas, dots, symbols
        words = text.split()
        ls.extend(words)
    else:
        ls.append('unknown')

df['video_tags'].apply(word_categorize)
```

```python
print(len(ls))      
print(ls[:30])      
```

```python
from collections import Counter

word_counts = Counter(ls)

# Show the top 20 most common words
print(word_counts.most_common(50))
```

```python
ls=[]
word_counts.most_common(1000)
for i in word_counts.most_common(1000):
    if len(i[0]) > 2:
      ls.append(i[0])
```

```python
stopwords = ['the', 'and', 'of', 'in', 'on', 'for', 'to', 'a', 'an', 'is', 'with']


ls = [word for word in ls if word not in stopwords]

print(ls[:50])  # preview first 50 words
```

```python
def video_tags_trend(tag):
    if isinstance(tag, str):
        tag = tag.lower()
        for word in ls:
            if word in tag:
                return word   
    return 'unknown' 


df['trend_tags'] = df['video_tags'].apply(video_tags_trend)


```

```python
from sklearn.ensemble import IsolationForest
num_df = df.select_dtypes(include=['int64', 'float64']).dropna()
iso = IsolationForest(contamination=0.05, random_state=42)
iso.fit(num_df)
df['outlier'] = iso.predict(num_df)
df_outliers = df[df['outlier'] == -1]
df_clean = df[df['outlier'] == 1]
print(f"Total detected outliers: {len(df_outliers)}")
print(f"Data shape after cleaning: {df_clean.shape}")
if len(num_df.columns) >= 2:
    x, y = num_df.columns[:2]
    plt.figure(figsize=(8,6))
    plt.scatter(df_clean[x], df_clean[y], label='Normal', alpha=0.6)
    plt.scatter(df_outliers[x], df_outliers[y], color='red', label='Outlier', alpha=0.7)
    plt.legend()
    plt.title('Isolation Forest Outlier Detection')
    plt.show()
```

```python
print(f' Num of removed rows = {df.shape[0] - df_clean.shape[0]}')
print(f' Num of kept rows    = {df.shape[0]}')
```
